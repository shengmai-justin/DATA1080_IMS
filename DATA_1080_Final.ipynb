````python

from langchain_community.chat_models import ChatOpenAI
from langchain.agents.agent_toolkits import SQLDatabaseToolkit
from langchain.utilities.sql_database import SQLDatabase
import autogen
import os
import time
import asyncio


class AutoReplyHandler:
    def __init__(self, timeout=10):
        self.timeout = timeout
        self.start_time = None
        self.escalation_levels = {
            "inventory_agent": ["coordinator_agent", "order_agent"],
            "order_agent": ["coordinator_agent", "task_agent"],
            "task_agent": ["coordinator_agent", "inventory_agent"]
        }
        
        # Define default actions for each agent
        self.default_actions = {
            "inventory_agent": self._default_inventory_action,
            "order_agent": self._default_order_action,
            "task_agent": self._default_task_action,
            "coordinator_agent": self._default_coordinator_action
        }
    
    def start_timer(self):
        self.start_time = time.time()
        
    def reset_timer(self):
        self.start_time = None
    
    def should_auto_reply(self):
        if not self.start_time:
            return False
        return time.time() - self.start_time > self.timeout

    def _default_inventory_action(self):
        return "SELECT * FROM Inventory WHERE quantity <= threshold"

    def _default_order_action(self):
        return "SELECT * FROM Orders WHERE status = 'pending' ORDER BY priority DESC LIMIT 5"

    def _default_task_action(self):
        return "SELECT * FROM Tasks WHERE status = 'pending' ORDER BY priority DESC LIMIT 5"

    def _default_coordinator_action(self):
        return "ESCALATE TO HUMAN"

    def handle_timeout(self, current_agent, all_agents, context=None):
        try:
            # Get the default action for the current agent
            default_action = self.default_actions.get(current_agent.name)
            
            if not default_action:
                return {"status": "error", "message": f"No default action defined for {current_agent.name}"}

            # Execute default action
            response = default_action()
            
            # If the action requires escalation
            if response == "ESCALATE TO HUMAN":
                return {
                    "status": "escalated",
                    "message": "Action requires human intervention",
                    "agent": current_agent.name
                }

            if current_agent.name in self.escalation_levels:
                next_agents = self.escalation_levels[current_agent.name]
                
                for next_agent_name in next_agents:
                    next_agent = next((a for a in all_agents if a.name == next_agent_name), None)
                    if next_agent:
                        return {
                            "status": "escalated",
                            "message": f"Escalated to {next_agent_name}",
                            "agent": next_agent_name,
                            "action": response
                        }

            return {
                "status": "completed",
                "action": response,
                "agent": current_agent.name
            }

        except Exception as e:
            return {
                "status": "error",
                "message": str(e),
                "agent": current_agent.name
            }

    def log_timeout_event(self, response):
        timestamp = time.strftime('%Y-%m-%d %H:%M:%S')
        print(f"[{timestamp}] AutoReply Event: {response}")

os.environ["OPENAI_API_KEY"] = "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"
database_path = 'warehouse.db'
db = SQLDatabase.from_uri(f'sqlite:///{database_path}')

llm = ChatOpenAI(model_name="gpt-4o", temperature=0)
toolkit = SQLDatabaseToolkit(db=db, llm=llm)
auto_reply_handler = AutoReplyHandler(timeout=10)



# Function Calling Schema setup
def generate_llm_config(tool):
    return {
        "name": tool.name.lower().replace(' ', '_'),
        "description": tool.description,
        "parameters": {
            "type": "object",
            "properties": tool.args if tool.args else {},
            "required": [],
        }
    }
tools = [generate_llm_config(tool) for tool in toolkit.get_tools()]
function_map = {tool.name: tool._run for tool in toolkit.get_tools()}
# AutoGen configuration
config_list = [{
    'model': 'gpt-4o',
    'api_key': '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'
}]

llm_config = {
    "functions": tools,
    "config_list": config_list,
    "timeout": 120,
}

# Add logging configuration
# import logging

# logging.basicConfig(level=logging.INFO)
# logger = logging.getLogger(__name__)

# # Define callback for agent responses
# def agent_response_callback(agent, response):
#     logger.info(f"Agent {agent.name} response: {response}")
````

````python
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
import time

class IMSChatInterface:
    def __init__(self, user_proxy, manager):
        self.user_proxy = user_proxy
        self.manager = manager
        
        # Style
        display(HTML("""
        <style>
        .chat-message { margin: 10px 0; padding: 10px; border-radius: 5px; }
        .user-message { background-color: #e3f2fd; }
        .agent-message { background-color: #f5f5f5; }
        .error-message { background-color: #ffebee; color: #c62828; }
        </style>
        """))
        
        # Create widgets
        self.output = widgets.Output(layout={
            'border': '1px solid #ddd',
            'min_height': '400px',
            'max_height': '400px',
            'overflow_y': 'auto',
            'margin': '10px 0'
        })
        
        self.text_input = widgets.Textarea(
            placeholder='Type your message here (Click Send or use the Send button)',
            layout={'width': '100%', 'height': '100px'}
        )
        
        self.send_button = widgets.Button(
            description='Send Message',
            button_style='primary',
            layout={'width': 'auto', 'margin': '10px 0'}
        )
        
        self.clear_button = widgets.Button(
            description='Clear Chat',
            button_style='warning',
            layout={'width': 'auto', 'margin': '10px 0'}
        )
        
        # Button container
        button_box = widgets.HBox([self.send_button, self.clear_button])
        
        # Main container
        self.container = widgets.VBox([
            widgets.HTML("<h2>IMS Multi-Agent Chat Interface</h2>"),
            self.output,
            self.text_input,
            button_box
        ])
        
        # Event handlers
        self.send_button.on_click(self.on_send)
        self.clear_button.on_click(self.on_clear)
        
        # Add keyboard shortcut for sending
        self.text_input.observe(self.handle_keyboard, names='value')
        
        # Display
        display(self.container)
    
    def handle_keyboard(self, change):
        # Check for Ctrl+Enter in the last keystroke
        if '\x0e' in change.new:  # Ctrl+Enter character
            # Remove the Ctrl+Enter character
            self.text_input.value = change.new.replace('\x0e', '')
            self.on_send(None)
    
    def on_clear(self, button):
        with self.output:
            clear_output()
    
    def format_message(self, sender, content, msg_type='normal'):
        if msg_type == 'user':
            return f'<div class="chat-message user-message"><strong>User:</strong><br>{content}</div>'
        elif msg_type == 'error':
            return f'<div class="chat-message error-message"><strong>Error:</strong><br>{content}</div>'
        else:
            return f'<div class="chat-message agent-message"><strong>{sender}:</strong><br>{content}</div>'
    
    def on_send(self, button):
        message = self.text_input.value.strip()
        if not message:
            return
            
        # Clear input
        self.text_input.value = ''
        
        with self.output:
            # Display user message
            display(HTML(self.format_message('User', message, 'user')))
            
            try:
                # Disable input while processing
                self.text_input.disabled = True
                self.send_button.disabled = True
                
                # Process with agents
                result = self.user_proxy.initiate_chat(
                    self.manager,
                    message=message
                )
                
                # Display agent responses
                if result.chat_history:
                    for msg in result.chat_history:
                        if msg.get('name') and msg.get('content'):
                            display(HTML(self.format_message(
                                msg['name'],
                                msg['content'].replace('\n', '<br>')
                            )))
                            time.sleep(0.3)  # Small delay for readability
                
            except Exception as e:
                display(HTML(self.format_message('System', str(e), 'error')))
            finally:
                # Re-enable input
                self.text_input.disabled = False
                self.send_button.disabled = False

# Sample test queries to display
test_queries = """
Example queries you can try (copy and paste):
1. "Check stock levels for all antibiotics"
2. "Process urgent order for Insulin"
3. "Show all pending tasks for Staff1"
4. "Update inventory for Aspirin"
5. "Check if any medicines are below their threshold levels"
"""

def create_chat_interface(user_proxy, manager):
    print("Test Queries for IMS:")
    print(test_queries)
    return IMSChatInterface(user_proxy, manager)
````

# Smart IMS GenAI System Documentation

## System Overview
The Smart IMS GenAI is a multi-agent system designed to optimize warehouse operations through intelligent inventory management, order processing, and task scheduling. The system implements the architecture outlined in the project proposal using SQLite for data persistence and AutoGen for agent coordination.

## Database Schema

### Inventory Table
```sql
CREATE TABLE Inventory (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id TEXT NOT NULL,
    name TEXT NOT NULL, 
    quantity INTEGER,
    location TEXT,
    threshold INTEGER,
    last_updated TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)
```

### Orders Table
```sql
CREATE TABLE Orders (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    order_id TEXT NOT NULL,
    product_id TEXT NOT NULL,
    quantity INTEGER,
    status TEXT,
    priority INTEGER,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    completed_at TIMESTAMP
)
```

### Tasks Table
```sql
CREATE TABLE Tasks (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    task_type TEXT,
    product_id TEXT,
    priority INTEGER,
    status TEXT,
    assigned_to TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    deadline TIMESTAMP
)
```

## Agent Implementation

### 1. Inventory Management Agent
- Monitors stock levels and thresholds
- Queries current inventory status
- Updates inventory quantities
- Triggers alerts for low stock
- SQL capabilities:
  - SELECT queries for stock checks
  - UPDATE statements for quantity adjustments
  - Monitoring threshold violations

### 2. Order Fulfillment Agent
- Processes new orders
- Manages order status updates
- Prioritizes order processing
- SQL capabilities:
  - INSERT statements for new orders
  - UPDATE statements for order status
  - SELECT queries for order tracking

### 3. Task Scheduling Agent
- Creates and assigns tasks
- Manages task priorities
- Updates task status
- SQL capabilities:
  - INSERT statements for new tasks
  - UPDATE statements for task status
  - SELECT queries for task monitoring
